```python
def get_df_each_from_files(directory, file_names):
    df = {}
    
    for file in file_names:
        video_id = get_video_id_from_file_name(file)
        df[video_id] = pd.read_json(directory + file, orient='records')
        df[video_id].set_index('id', inplace=True)
    
    return df


df_each = get_df_each_from_files(directory, files)
print(len(df_each))
```

# POS

```python
def get_pos_from_comment(df, tagger_name='mecab'):
    tagger_lower = tagger_name.lower()
    if tagger_lower == 'mecab':
        from konlpy.tag import Mecab
        tagger = Mecab()
    elif tagger_lower == 'kkma':
        from konlpy.tag import Kkma
        tagger = Kkma()
    elif tagger_lower == 'okt':
        from konlpy.tag import Okt
        tagger = Okt()

    lines = ' '.join([t for t in df['commentText'].tolist() if t != ''])
    return tagger.pos(lines)


def get_all_tag_set_from_pos(pos):
    return set(tuple(zip(*pos))[1])


def get_n_words_for_tag_from_pos(pos, tag, count=None):
    words = tuple(p[0] for p in pos if p[1]==tag)
    if count:
        return words[:count]
    else:
        return words


def get_counter_from_words(words):
    from collections import Counter
    return Counter(words)


def get_counter_from_df(df, tag, tagger_name='mecab'):
    pos = get_pos_from_comment(df, tagger_name)
    words = get_n_words_for_tag_from_pos(pos, tag)
    return get_counter_from_words(words)


def get_counter_merged_from_df(df, tag, tagger_name='mecab'):
    from collections import Counter
    interval = 30000
    total = len(df)
    # interval + 1: in case total=30000
    total_cycle = (total // (interval + 1)) + 1
    counter_merged = Counter()
    for cycle in range(total_cycle):
        start = cycle * interval
        end = ((cycle + 1) * interval) + 1
        df_piece = df[start:end]
        counter_merged = counter_merged + get_counter_from_df(df_piece, tag, tagger_name=tagger_name)
    return counter_merged


def get_df_from_keyword(df, keyword_list):
    import pandas as pd
    idx_all_keywords = [True] * len(df)
    
    if isinstance(keyword_list, str):
        keyword_list = [keyword_list]

    for keyword in keyword_list:
        idx_all_keywords = idx_all_keywords & df['commentText'].str.contains(keyword)

    return df.loc[idx_all_keywords, ['commentText']]


def get_counter_from_df_by_keyword_and_tag(df, keyword_list, tag):
    df_keyword = get_df_from_keyword(df, keyword_list)
    return get_counter_merged_from_df(df_keyword, tag)


def get_suffix_added_counter_from_counter(counter):
    from collections import Counter
    return Counter({k + '다': v for k, v in counter.items()})


def get_suffix_added_counter_from_df(df, tag, tagger_name='mecab'):
    cnt = get_counter_merged_from_df(df, tag, tagger_name)
    return get_suffix_added_counter_from_counter(cnt)


def get_suffix_added_counter_from_df_by_keyword_and_tag(df, keyword, tag):
    cnt = get_counter_from_df_by_keyword_and_tag(df, keyword, tag)
    return get_suffix_added_counter_from_counter(cnt)


def print_len_and_body(data):
    print(len(data))
    print(data)


def export_df(df, file_path, sep='\t', index=False, reset_index=False, drop=False):
    if reset_index:
        df.reset_index(drop=drop, inplace=True)
    df.to_csv(file_path, sep=sep, index=index)


def export_counter(counter, file_path, n=10, sep='\t', index=False):
    df_cnt = pd.DataFrame(columns=['key', '횟수'], data=counter.most_common(n))
    export_df(df_cnt, file_path, sep=sep, index=index)
```

```python
def get_comment_from_user(df, user, head=5):
    idx_user = df['user']==user
    cols = ['commentText']
    return df.loc[idx_user, cols].head(head)


def get_df_from_user(df, user):
    idx_user = df['user']==user
    return df.loc[idx_user]


def get_video_id_from_comment_id(df_dict, _id):
    video_id = None
    for vid, df in df_dict.items():
        if _id in df.index:
            video_id = vid
            break
    return video_id


def get_video_url_from_comment_id(df_dict, _id):
    video_id = get_video_id_from_comment_id(df_dict, _id)
    if video_id:
        return f'https://www.youtube.com/watch?v={video_id}'
    else:
        return None
```

# Replies

```python
def get_replies_from_df(df):
    from functools import reduce
    idx_exists = ~df['replies'].isna()
    return reduce(lambda x, y: x + y, df.loc[idx_exists, 'replies'], [])


def get_df_from_list(_list):
    return pd.DataFrame(data=_list).set_index('id')


def get_df_replies_from_df(df):
    replies = get_replies_from_df(df)
    return get_df_from_list(replies)


def get_top_n_replies_from_df(df, n=10):
    return df.nlargest(n, 'numberOfReplies')


def get_mean_number_of_replies_from_df(df):
    return df['numberOfReplies'].mean()


def get_sentences_including_keyword_from_df(df, keyword):
    idx_include_keyword = df['commentText'].str.contains(keyword)
    return df.loc[idx_include_keyword]


def get_number_of_has_replies_from_df(df):
    return df.loc[df['hasReplies'], 'hasReplies'].count()


def get_replies_by_id(df, _id):
    return df.loc[df.index.str.contains(_id)]
```

# Http

```python
def get_http_list_from_df(df):
    import re
    total_comment = ' '.join(df['commentText'].tolist())
    return re.findall("(https?://[^\s]+)", total_comment)


def get_netloc_counter_from_set(_set):
    from urllib.parse import urlsplit
    from collections import Counter
    netloc_list = [urlsplit(s)[1] for s in _set]
    return Counter(netloc_list)


def get_url_counter_from_list(_list):
    from urllib.parse import urlsplit
    from collections import Counter
#     netloc_list = [urlsplit(l)[1] for l in _list]
    return Counter(_list)


def get_netloc_counter_merged_from_set(_set):
    from urllib.parse import urlsplit
    from collections import Counter
    
    def get_netloc_merged_from_str(netloc):
        if 'naver' in netloc:
            return 'naver'
        elif ('youtu.be' in netloc) or ('youtube' in netloc):
            return 'youtube'
    
    netloc_list = [urlsplit(s)[1] for s in _set]
    
```
