# Functions

```python
def get_raw_from_file(file_name):
    directory = '/home/jake/Documents/ALIO/'
    with open(directory + file_name, encoding='euckr') as f:
        head = f.readline()
        head = head.replace(' ', '')
        head = head.replace('2018단위', '2018,단위')
        body = f.read()
    raw = head + body
    return raw

def get_df_from_raw(raw, sep=','):
    from io import StringIO
    sio = StringIO(raw)
    df = pd.read_csv(sio, sep=sep)
    df['항목'] = df['항목'].apply(str.strip)
    return df

def get_df_from_file(file, sep=','):
    raw = get_raw_from_file(file)
    df = get_df_from_raw(raw, sep=sep)
    return df
```

## Pivot

```python
df_rcrt_pivot = df_rcrt.loc[idx_rcrt].pivot(index='기관명', columns='항목', values=col_year)
df_rcrt_pivot.columns.set_levels(['신입_지역인재', '신입_여성', '신입_전체', '직원_전체', '인턴_채용형', '인턴_체험형'], level=1, inplace=True)
df_rcrt_pivot
```

```python
df_rcrt_pivot.loc[~df_rcrt_pivot.index.str.contains('병원'), [('2018', '신입_여성')]].sort_values(('2018', '신입_여성'), ascending=False).head(20)
```

```python
'''
체험인턴 / 직원전체
채용인턴 / 직원전체
신입전체 / 직원전체
여성 / 신입전체
지역인재 / 신입전체
'''

for y in col_year:
    df_rcrt_pivot[(y, '체험인턴/직원전체')] = df_rcrt_pivot[(y, '인턴_체험형')] / df_rcrt_pivot[(y, '직원_전체')] * 100
    df_rcrt_pivot[(y, '채용인턴/직원전체')] = df_rcrt_pivot[(y, '인턴_채용형')] / df_rcrt_pivot[(y, '직원_전체')] * 100
    df_rcrt_pivot[(y, '신입전체/직원전체')] = df_rcrt_pivot[(y, '신입_전체')] / df_rcrt_pivot[(y, '직원_전체')] * 100
    df_rcrt_pivot[(y, '여성/신입전체')] = df_rcrt_pivot[(y, '신입_여성')] / df_rcrt_pivot[(y, '신입_전체')] * 100
    df_rcrt_pivot[(y, '지역인재/신입전체')] = df_rcrt_pivot[(y, '신입_지역인재')] / df_rcrt_pivot[(y, '신입_전체')] * 100

df_rcrt_pivot.head()
# df_pay_pivot_woman.sort_values(('2018', '여성 비율'), ascending=True)
```
