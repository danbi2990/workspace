# Functions

```python
def get_raw_from_file(file_name):
    directory = '/home/jake/Documents/ALIO/'
    with open(directory + file_name, encoding='euckr') as f:
        head = f.readline()
        head = head.replace(' ', '')
        head = head.replace('2018단위', '2018,단위')
        body = f.read()
    raw = head + body
    return raw

def get_df_from_raw(raw, sep=','):
    from io import StringIO
    sio = StringIO(raw)
    df = pd.read_csv(sio, sep=sep)
    df['항목'] = df['항목'].apply(str.strip)
    return df

def get_df_from_file(file, sep=','):
    raw = get_raw_from_file(file)
    df = get_df_from_raw(raw, sep=sep)
    return df
```
