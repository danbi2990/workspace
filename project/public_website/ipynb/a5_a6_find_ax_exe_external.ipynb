```python
import re
from pprint import pprint
from collections import defaultdict

import pandas as pd

from db.mongo import MyMongo
from project.public_website.ipynb.jupyter_functions import get_comment, extend_exe, remove_comment, dict2df


with MyMongo() as db:
    js_files = db.find('public_website', 'a5_js_code_external')

def remove_comment(script):
    script = re.sub(r'(\/\/.*\n)', '', script)
    script = re.sub(r'(\/\*[\s\S]*?\*\/)', '', script)
    return script


def get_comment(script):
    one_line = re.findall(r'(\/\/.*\n)', script)
    multi_line = re.findall(r'(\/\*[\s\S]*?\*\/)', script)

    return '\n'.join(one_line) + '\n' + '\n'.join(multi_line)


def get_cab(script):
    return re.findall(r'\/(\S*\.cab)', script)


def extend_cab(cab, script):
    new_cab = get_cab(script)
    if new_cab:
        cab.extend(new_cab)


res = []
i = 0
for u in js_files:
#     if i == 500:
#         break
#     i += 1
    
    url = u['webPath']
    script = u['jsSource']
    cmt = get_comment(script)
    code = remove_comment(script)
    cab_cmt = set(re.findall(r'\/(\S*\.cab)', cmt))
    cab_code = set(re.findall(r'\/(\S*\.cab)', code))
    exe_cmt = set(re.findall(r'\/(\S*\.exe)[^c]', cmt))
    exe_code = set(re.findall(r'\/(\S*\.exe)[^c]', code))
    
    if cab_cmt or cab_code or exe_cmt or exe_code:
        res.append((url, cab_cmt, cab_code, exe_cmt, exe_code))


pprint(res)
```

```python
df_0 = pd.DataFrame(columns=['url', 'ax_cmt', 'ax_code', 'exe_cmt', 'exe_code'], data=res)
df_0[['ax_cmt', 'ax_code', 'exe_cmt', 'exe_code']] = df_0[['ax_cmt', 'ax_code', 'exe_cmt', 'exe_code']].applymap(lambda x: list(x))
```
