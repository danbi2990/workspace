# Fetch external

```python
with MyMongo() as db:
    external = db.get_df_from_table('public_website', 'a6_ax_exe_external')
```

```python
external.head()
```

```python
pd.set_option('display.max_rows', 30)
# 13817
ax_srt = external.loc[external['ax_srt']!=''][['url', 'ax_srt']]
ax_srt['net_loc'] = ax_srt['url'].apply(lambda x: urlsplit(x)[1])
ax_srt['path'] = ax_srt['url'].apply(lambda x: urlsplit(x)[2]); len(ax_srt)

# _0x980b[195]: 난독화
# shockwaveflash => ShockwaveFlash
# xmlhttp
# xmldom
# scripting.filesystemobject
# variables: c, j, version[data]...
# htmlfile: html doc 생성, createElement 등 html 함수 사용. 메모리 관리하지 않으면 릭 발생

```

```python
# ax_srt.loc[ax_srt['ax_srt'].str.contains('Shockwave')].at['ax_srt'] = 'shockwaveflash'
ax_srt['ax_srt'].replace(to_replace=r'.*Shockwave.*', value='shockwaveflash', inplace=True, regex=True)
ax_srt['ax_srt'].replace(to_replace=r'.*SHOCKWAVE.*', value='shockwaveflash', inplace=True, regex=True)

ax_srt['ax_srt'].replace(to_replace=r'.*XMLDOM.*', value='xmldom', inplace=True, regex=True)
ax_srt['ax_srt'].replace(to_replace=r'.*XmlDom.*', value='xmldom', inplace=True, regex=True)
ax_srt['ax_srt'].replace(to_replace=r'.*Msxml2.*', value='xmldom', inplace=True, regex=True)
ax_srt['ax_srt'].replace(to_replace=r'.*MSXML2.*', value='xmldom', inplace=True, regex=True)
ax_srt['ax_srt'].replace(to_replace=r'.*msxml.*', value='xmldom', inplace=True, regex=True)


# progIDs[i]: Msxml.dom

ax_srt['ax_srt'].replace(to_replace=r'.*XMLHTTP.*', value='xmlhttp', inplace=True, regex=True)
ax_srt['ax_srt'].replace(to_replace=r'.*XMLHttp.*', value='xmlhttp', inplace=True, regex=True)
ax_srt['ax_srt'].replace(to_replace=r'navigator.userAgent.indexOf\(\'MSIE 5', value='xmlhttp', inplace=True, regex=True)


# axarray[i] in 119.seoul.go.kr/dwr/engine.js : arbitrary ax
# dwr.engine.serialize.domDocument, dwr.engine._DOMDocument: DOM
# AgControl.AgControl: silverlight
# stlib.stfp.checkIEPlugins[a] in http://argentina.korean-culture.org/js/snsShare.js: tracker
# AxKSignNTCA.AxKSignNTCA.1 in ksign_script.js
# _0x980b[194] in xkeypad_desktop.js
# _0x980b[195] in xkeypad_desktop.js
# _0x980b[196] in xkeypad_desktop.js
```

```python
pd.set_option('display.max_rows', None)

categorized = ['xmlhttp', 'xmldom', 'htmlfile', 'shockwaveflash', 'Scripting.FileSystemObject', 'axarray[i]', 'dwr.engine.serialize.domDocument',
               'dwr.engine._DOMDocument' ,'AgControl.AgControl', 'stlib.stfp.checkIEPlugins[a]', 'AxKSignNTCA.AxKSignNTCA.1 in ksign_script.js',
               'AxKSignNTCA.AxKSignNTCA.1', '_0x980b[194]', '_0x980b[195]', '_0x980b[196]']
ax_srt_rm = ax_srt.loc[~ax_srt['ax_srt'].isin(categorized)]
ax_srt_rm.groupby(['net_loc', 'path', 'ax_srt']).agg('count')
```
