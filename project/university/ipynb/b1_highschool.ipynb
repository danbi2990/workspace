# Functions

```python
def convert_str2int(d, idx_numeric):
    d.iloc[::,idx_numeric:] = d.iloc[::,idx_numeric:].applymap(lambda x: pd.to_numeric(x.replace(',', '') if type(x)!=int and type(x)!=float else x))
```

## 2016, 17, 18 출신 고등학교 비율

```python
tmp = pd.DataFrame(data=df['2018'].loc[idx_2018, cols_count].sum().astype(int), columns=['2018'])
tmp2 = pd.DataFrame(data=df['2017'].loc[idx_2017, cols_count].sum().astype(int), columns=['2017'])
tmp3 = pd.DataFrame(data=df['2016'].loc[idx_2016, cols_count].sum().astype(int), columns=['2016'])

tmp4 = tmp3.merge(tmp2, left_index=True, right_index=True, how='outer')
tmp5 = tmp4.merge(tmp, left_index=True, right_index=True, how='outer')

tmp6 = tmp5.reset_index()
tmp7 = tmp6.pivot_table(columns='index', values=['2016', '2017', '2018'])

del tmp, tmp2, tmp3, tmp4, tmp5, tmp6

tmp7['일반고 비율'] = tmp7['일반고 학생수'] / tmp7['총입학자수'] * 100

tmp7['특목고 비율'] = tmp7['특목고 학생수'] / tmp7['총입학자수'] * 100

tmp7['과학고 비율'] = tmp7['과학고 학생수'] / tmp7['총입학자수'] * 100
tmp7['예술· 체육고 비율'] = tmp7['예술· 체육고 학생수'] / tmp7['총입학자수'] * 100
tmp7['외고· 국제고 비율'] = tmp7['외고· 국제고 학생수'] / tmp7['총입학자수'] * 100
tmp7['특성화고 비율'] = tmp7['특성화고 학생수'] / tmp7['총입학자수'] * 100

tmp7['산업수요맞춤형 고등학교 비율'] = tmp7['산업수요맞춤형 고등학교 학생수'] / tmp7['총입학자수'] * 100
tmp7['자율고 비율'] = tmp7['자율고 학생수'] / tmp7['총입학자수'] * 100
tmp7['영재학교 비율'] = tmp7['영재학교 학생수'] / tmp7['총입학자수'] * 100

tmp7['검정고시 비율'] = tmp7['검정고시 학생수'] / tmp7['총입학자수'] * 100
tmp7['그 외 기타 비율'] = tmp7['그 외 기타 학생수'] / tmp7['총입학자수'] * 100
tmp7 = tmp7.reset_index()
tmp7 = tmp7.rename(columns={'index':'연도'})

cols_rate_without_school = ['연도', *cols_rate]
cols_rate_without_school.remove('학교')

tmp7.loc[::, cols_rate_without_school]


# tmp = tmp.pivot_table(columns='year', values='2018'); tmp
```
